```python
import pandas as pd

# Cargar los ficheros de balance
path_bal_general = "data/us-balance-quarterly.csv"
path_bal_banks = "data/us-balance-banks-quarterly.csv"
path_bal_insurance = "data/us-balance-insurance-quarterly.csv"

df_general = pd.read_csv(path_bal_general, sep=';')
df_banks = pd.read_csv(path_bal_banks, sep=';')
df_insurance = pd.read_csv(path_bal_insurance, sep=';')

df_general.shape, df_banks.shape, df_insurance.shape
```

```python
print("Tickers en df general:",df_general['Ticker'].nunique())
print("Tickers en df banks:",df_banks['Ticker'].nunique())
print("Tickers en df insurance:",df_insurance['Ticker'].nunique())
```

```python
# Cargar Tickers del S&P 500
from src.funcionesExtract import limpieza_tickers, clean_ticker
path_sp500 = "data/constituents.csv" 
df_sp500 = pd.read_csv(path_sp500)
sp500_clean = limpieza_tickers(df_sp500)
tickers_sp = set(sp500_clean['Ticker'])

print("Tickers en S&P:", len(tickers_sp))
```

```python
# Limpiar tickers y sets de valores unicos
df_general['Ticker_clean'] = df_general['Ticker'].map(clean_ticker)
df_banks['Ticker_clean'] = df_banks['Ticker'].map(clean_ticker)
df_insurance['Ticker_clean'] = df_insurance['Ticker'].map(clean_ticker)

tickers_general = set(df_general['Ticker_clean'].unique())
tickers_banks = set(df_banks['Ticker_clean'].unique())
tickers_insurance = set(df_insurance['Ticker_clean'].unique())

len(tickers_general), len(tickers_banks), len(tickers_insurance)
```

```python
# Verificar si hay tickers repetidos
len(tickers_general & tickers_banks), len(tickers_general & tickers_insurance), len(tickers_insurance & tickers_banks)
```

```python
# Tickers del indice s&p en cada uno
len(tickers_general & tickers_sp), len(tickers_banks & tickers_sp), len(tickers_insurance & tickers_sp)
```

```python
df_banks.groupby('Ticker_clean')['Fiscal Year'].nunique().describe()
```

```python
# unir los 3 sets de tickers y ver missings con respecto al indice
tickers_simfin = tickers_general | tickers_banks | tickers_insurance
missing = sorted(tickers_sp - tickers_simfin)
missing[:], len(missing)
```
